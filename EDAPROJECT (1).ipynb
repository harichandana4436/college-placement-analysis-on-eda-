```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("CollegePlacement.csv")

# Select only numeric columns
numeric_cols = data.select_dtypes(include=['number']).columns

# Set up the plotting grid
plt.figure(figsize=(15, 8))

# Create a boxplot for each numeric column
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, (len(numeric_cols) + 1) // 2, i)
    sns.boxplot(y=data[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

- by using the median we have fill the outliers

```python
import pandas as pd
import numpy as np

# Assuming you already have a DataFrame 'data' with column 'IQ'

# Calculate Q1, Q3, and IQR for 'IQ'
Q1 = data['IQ'].quantile(0.25)
Q3 = data['IQ'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Compute median of non-outlier values
median_IQ = data.loc[(data['IQ'] >= lower_bound) & (data['IQ'] <= upper_bound), 'IQ'].median()

# Replace outliers with median
data['IQ'] = np.where((data['IQ'] < lower_bound) | (data['IQ'] > upper_bound),
                      median_IQ,
                      data['IQ'])
```

```python
import seaborn as sns

plt.figure(figsize=(5, 5))
sns.boxplot(y=data['IQ'], color='skyblue')
plt.title('Box Plot of IQ')
plt.ylabel('IQ')
plt.show()
```
